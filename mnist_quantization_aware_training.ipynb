```python
import lightning as L
import matplotlib.pyplot as plt
import torch
from lightning import seed_everything
from lightning.pytorch.callbacks import EarlyStopping
from lightning.pytorch.tuner import Tuner
from torchvision import datasets
from torchvision import transforms

from src.module import ResNet, ResidualBlock

seed_everything(42, workers=True)

%matplotlib inline
```

### Training with LearningRateFinder.

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
torch.set_default_device(device)
```

```python
trainer = L.Trainer(
    max_epochs=20,
    callbacks=[EarlyStopping(monitor="val_acc", min_delta=1e-2, patience=5, verbose=True, mode='min')],
)
tuner = Tuner(trainer)
```

```python
model = ResNet(block=ResidualBlock, layers=[3, 4, 6, 3], device=device.type, quantize=True)
model.to(device)
device
```

```python
# tuner.scale_batch_size(model, init_val=2 ** 15)
```

```python
lr_finder = tuner.lr_find(model)

lr_finder.plot(suggest=True)
```

```python
del lr_finder, tuner
```

```python
model.hparams.batch_size, model.hparams.learning_rate
```

```python
trainer.fit(model)
```

```python
trainer.test(model)
```

```python
# Conversion after training
quantized_model = torch.quantization.convert(model.eval(), inplace=False)
```
